```python
def agregar_keys(iterador, padre=0, codigo=''):
    count = 1
    for nodo in iterador:
        if nodo['data']['padre'] == 0:
            key = str(count)
            nodo['data']['cod_ext'] = nodo['data']['codigo']
        else:
            key = f'{padre}.{count}'
            nodo['data']['cod_ext'] = f"{codigo}.{nodo['data']['codigo']}"
            
        nodo['key'] = key

        if len(nodo['children']):
            agregar_keys(nodo['children'], padre=key, codigo=nodo['data']['cod_ext'])

        count += 1

def generar_treenodes(lista: list) -> dict:

    nodos = []

    # Se le da el formato necesario del nodo
    for i in lista:
        nodos.append({ 'key': None, 'data': i, 'children': [] })

    # Se ordenan de mayor a menor:
    nodos = sorted(nodos, key=lambda d: -d['data']['padre']) 

    # Se incrustan los hijos dentro de los padres:
    for nodo in nodos.copy():
        if nodo['data']['padre'] != 0:
            padre = next(item for item in nodos if item['data']['id'] == nodo['data']['padre'])
            padre['children'].append(nodo)
            nodos.remove(nodo)

    # Se agregan las keys:
    agregar_keys(nodos)

    # Se devuelve el dict de salida
    return {'root': nodos}
```
